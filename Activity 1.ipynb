# Activity 1: Transform XML to data frame to CSV.

Loading requires libraries :

```r
library(XML)
library(plyr)
library(dplyr)
options(warn=-1)
```

Loading and Parsing XML data and converting it to a list :

```r
fileurl <- "http://aiweb.cs.washington.edu/research/projects/xmltk/xmldata/data/courses/reed.xml"
data <- xmlParse(fileurl, useInternalNodes = TRUE)
xml <- xmlToList(data, addAttributes = TRUE, simplify = FALSE)
```

1) Converting first 8 columns of list to data frame by seperating out each rows of 9 columns and using rbind <br>
2) Seperating 9th column of time similarly and replacing NULL values by NA (done so that unlist retains all the rows) and rbind, making a new data frame. <br>
3) Similarly seperating 10th column and making a new data frame.

```r
#1)
xml_data <- lapply(xml, "[",c(1:8))
df <-data.frame(do.call("rbind", xml_data))

#2)
xml_data <- lapply(xml, "[",c(9))
xml_data <- do.call(rbind, lapply(xml_data, rbind))
xml_data <- lapply(xml_data, function(x) ifelse(x == "NULL", NA, x))
df1 <-data.frame(do.call("rbind", lapply(xml_data, unlist)))

#3)
xml_data <- lapply(xml, "[",c(10))
df2 <-data.frame(do.call("rbind", lapply(xml_data, unlist)))
```

Binding all data frames together using cbind :

```r
df <- cbind(df, df1)
df <- cbind(df, df2)
df
```

Writing data frame to csv file :

```r
df = as.matrix(df)
write.csv(df,"reed.csv")
```
